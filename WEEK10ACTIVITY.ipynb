# 4. Carry out correlation analysis on the dataset and find the correlation factor between                
a. ph and Chloramines                                                                                                
b. ph and Conductivity

```python
corrmatrix=data.corr()
plt.figure(figsize=(10,6))
sns.heatmap(corrmatrix,vmin=0.2,vmax=0.9,annot=True,linewidth=0.2,cmap='BuPu')
```

a. correlation between ph and Chloramines

```python
df1=data[['ph','Chloramines']]
corrmatrix1=df1.corr()
plt.subplots(figsize=(8,8))
sns.heatmap(corrmatrix1,vmin=.2,vmax=1,annot=True,linewidth=.2,cmap='Greens')
```

correlation factor between ph and Chloramines is -0.025 which is near to 0 . therefore ph and Chloramines has small negative correlation

b. correlation between ph and Conductivity

```python
df2=data[['ph','Conductivity']]
corrmatrix2=df2.corr()
plt.subplots(figsize=(8,8))
sns.heatmap(corrmatrix2,vmin=.2,vmax=1,annot=True,linewidth=.2,cmap='Blues')
```

correlation factor between ph and Conductivity is 0.014 which is near to 0 . therefore ph and Conductivity has small positive correlation
